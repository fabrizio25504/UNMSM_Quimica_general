# **Propiedades de los l√≠quidos**

image.png

Por qu√© los liquidos estan mucho mas compactos a diferencia de los gases y adem√°s sus moleculas interactuan constantemente(fuerzas intermoleculares) a diferencia de los gases.

image.png

Esto se debe a las fuerzas intermoleculares m√°s especificamente la tensi√≥n superficial.
Gracias a estas fuerzas las moleculas al borden del vaso son atraidas por las moleculas debajo de ellas y no solo sino tambien con el apoyo que generan estas moleculas con las moleculas del vaso(adhesi√≥n).

image.png

La viscosidad es la resistencia de un l√≠quido a fluir, y es una propiedad que est√° relacionada con las fuerzas intermoleculares en el l√≠quido.

 La viscosidad es una medida de la fricci√≥n interna en un l√≠quido que se opone al flujo de las mol√©culas. Cuanto m√°s fuertes sean las fuerzas intermoleculares en un l√≠quido, mayor ser√° su viscosidad. Esta relaci√≥n implica que la viscosidad puede variar con la temperatura, ya que las fuerzas intermoleculares suelen cambiar con la temperatura.

image.png

Lo que pasa es que en el agua la energ√≠a cinetica permite que las moleculas de agua constante rompan y formen enlaces de hidrogeno, pero cuando la temperatura baja lo suficiente la energ√≠a cinetica sera menor que la fuerza de los enlaces de hidrogeno por lo que comenzaran a formar una estructura cristalina donde hay espacios entre si creando un mayor volumen con la misma cantidad de sustancia.
