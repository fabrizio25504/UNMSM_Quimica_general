## Pregunta 2: Configuración electrónica (4 puntos)

Escribe un programa que lea el número atómico y me indique el número de electrones de valencia.

```python
# Número de protones como input
numero_atomico = int(input("ingresa el número de protones(del 1 al 86): "))
Sharp=2
Principal=6
Difusse=10
Fundamental=14
valencia = 0
def valencia_function(S,P,D,F,numero_atomico):
  if S == 1 :
    if numero_atomico <= 2 :
      valencia = numero_atomico
      print(f"la valencia es :{valencia}")
      return valencia
    numero_atomico = numero_atomico - Sharp
  if F == 1 :
    numero_atomico = numero_atomico - Fundamental
    if numero_atomico <= 0 :
      valencia = Sharp
      print(f"la valencia es :{valencia}")
      return valencia
  if D == 1 :
    numero_atomico = numero_atomico - Difusse
    if numero_atomico <= 0 :
      valencia = Sharp
      print(f"la valencia es :{valencia}")
      return valencia
  if P == 1 :
    valencia = numero_atomico
    valencia = valencia + Sharp
    print(f"la valencia es :{valencia}")
    return valencia


if numero_atomico <= 2 :
  valencia = numero_atomico
  print(f"la valencia es :{valencia}")
elif numero_atomico > 2 and numero_atomico <= 10 :
  valencia=numero_atomico-2
  print(f"la valencia es :{valencia}")
elif numero_atomico > 10 and numero_atomico <= 18:
  valencia = numero_atomico-10
  print(f"la valencia es :{valencia}")
elif numero_atomico > 18 and numero_atomico <= 36:
  numero_atomico = numero_atomico-18
  valencia_function(1,1,1,0,numero_atomico)
elif numero_atomico > 36 and numero_atomico <= 54:
  numero_atomico = numero_atomico-36
  valencia_function(1,1,1,0,numero_atomico)
elif numero_atomico > 54 and numero_atomico <= 86:
  numero_atomico = numero_atomico-54
  valencia_function(1,1,1,1,numero_atomico)
```

```python
print("REACCIÓN DE COMBUSTIÓN DE")
print('--------------------------------')
print('    1. Alcanos\n    2. Alquenos\n    3. Alquinos')
print('--------------------------------')

rxn = int(input('Indica el tipo de hidrocarburo: 1, 2 o 3: '))
while (rxn != 1) and (rxn != 2) and (rxn != 3):
    print("¡Ha escrito un número inválido! Inténtelo de nuevo")
    rxn = int(input('Indica el tipo de hidrocarburo: 1, 2 o 3: '))

#**********************************************************************************
#función para usar subindice
def get_sub(x):
    normal = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789+-=()"
    sub_s = "ₐ₈CDₑբGₕᵢⱼₖₗₘₙₒₚQᵣₛₜᵤᵥwₓᵧZₐ♭꜀ᑯₑբ₉ₕᵢⱼₖₗₘₙₒₚ૧ᵣₛₜᵤᵥwₓᵧ₂₀₁₂₃₄₅₆₇₈₉₊₋₌₍₎"
    res = x.maketrans(''.join(normal), ''.join(sub_s))
    return x.translate(res)
#**********************************************************************************

if rxn == 1:
    print(' ')
    print('[Combustión de Alcanos]')
    print('a C{}H{} + b O{} = c CO{} + d H{}O'.format(get_sub('n'),get_sub('2n+2'),get_sub('2'),get_sub('2'),get_sub('2')))
elif rxn == 2:
    print(' ')
    print('[Combustión de Alquenos]')
    print('a C{}H{} + b O{} = c CO{} + d H{}O'.format(get_sub('n'),get_sub('2n'),get_sub('2'),get_sub('2'),get_sub('2')))
elif rxn == 3:
    print(' ')
    print('[Combustión de Alquinos]')
    print('a C{}H{} + b O{} = c CO{} + d H{}O'.format(get_sub('n'),get_sub('2n-2'),get_sub('2'),get_sub('2'),get_sub('2')))
```

```python
if rxn == 1:
    carbonos = int(input('Indica el número de carbonos en el alcano: '))
    a = 1
    b = (3*carbonos+1)/2
    c = carbonos
    d = carbonos+1
    print(' ')
    print('---ECUACIÓN BALANCEADA---')
    print('C{}H{} + {} O{} = {} CO{} + {} H{}O'.
          format(get_sub(str(carbonos)),get_sub(str(2*carbonos+2)),str(b),
                 get_sub('2'),str(c),get_sub('2'),str(d),get_sub('2')))
    print("la molecularidad de los reactantes: ", a+b)
    print("la molecularidad del producto es: ", c+d)

if rxn == 2:
    carbonos = int(input('Indica el número de carbonos en el alqueno: '))
    a = 1
    b = (3*carbonos)/2
    c = carbonos
    d = carbonos
    print(' ')
    print('---ECUACIÓN BALANCEADA---')
    print('C{}H{} + {} O{} = {} CO{} + {} H{}O'.
          format(get_sub(str(carbonos)),get_sub(str(2*carbonos)),str(b),
                 get_sub('2'),str(c),get_sub('2'),str(d),get_sub('2')))
    print("la molecularidad de los reactantes: ", a+b)
    print("la molecularidad del producto es: ", c+d)
if rxn == 3:
    carbonos = int(input('Indica el número de carbonos en el alquino: '))
    a = 1
    b = (3*carbonos-1)/2
    c = carbonos
    d = carbonos-1
    print(' ')
    print('---ECUACIÓN BALANCEADA---')
    print('C{}H{} + {} O{} = {} CO{} + {} H{}O'.
          format(get_sub(str(carbonos)),get_sub(str(2*carbonos-2)),str(b),
                 get_sub('2'),str(c),get_sub('2'),str(d),get_sub('2')))
    print("la molecularidad de los reactantes: ", a+b)
    print("la molecularidad del producto es: ", c+d)
```
